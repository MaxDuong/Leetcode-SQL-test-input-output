```python
import json
import pandas as pd
```

# INPUT

```python
file = 'input.json'
with open(file) as json_file:
    dict_input = json.load(json_file)

print("A list of tables' name")
for i in dict_input['headers'].keys():
    print(i)
```

```python
def show_table_input(table_name):
    file = 'input.json'
    with open(file) as json_file:
        dict_input = json.load(json_file)
    
    return pd.DataFrame(columns=dict_input['headers'][table_name], 
             data=dict_input['rows'][table_name])
```

```python
table_name = 'Listens'
show_table_input(table_name)
```

```python
df = show_table_input(table_name)
df[df['user_id'] == 1].sort_values(by=['day'])[:20]
```
